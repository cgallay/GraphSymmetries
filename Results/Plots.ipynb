```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline 
```

# Experience 1
Look at the influance of the size of the train set for both the CNN model and the Gaph Convolution model.

```python
df_result = pd.DataFrame(columns=['Accuracy', 'Type', 'nb_param', 'Dataset', 'Remark'])
df_result = df_result.append({'Accuracy': 82.30,
                              'Type': 'Graph',
                              'nb_param': 987140,
                              'Dataset':'AID',
                              'Train_size': 0.8,
                              'Remark': 'Size before FC is 5x5, kernel_size=5'}, ignore_index=True)

df_result = df_result.append({'Accuracy': 80.20,
                              'Type': 'CNN',
                              'nb_param': 1150852,
                              'Dataset':'AID',
                              'Train_size': 0.8,
                              'Remark': 'Kernel_size=(3,3)'}, ignore_index=True)

df_result = df_result.append({'Accuracy': 60.0,
                              'Type': 'Graph',
                              'nb_param': 987140,
                              'Dataset':'AID',
                              'Train_size': 0.2,
                              'Remark': 'Size before FC is 5x5, kernel_size=5'}, ignore_index=True)

df_result = df_result.append({'Accuracy': 50.0,
                              'Type': 'CNN',
                              'nb_param': 1150852,
                              'Dataset':'AID',
                              'Train_size': 0.2,
                              'Remark': 'Kernel_size=(3,3)'}, ignore_index=True)

df_result = df_result.append({'Accuracy': 75.0,
                              'Type': 'Graph',
                              'nb_param': 987140,
                              'Dataset':'AID',
                              'Train_size': 0.5,
                              'Remark': 'Size before FC is 5x5, kernel_size=5'}, ignore_index=True)

df_result = df_result.append({'Accuracy': 67.0,
                              'Type': 'CNN',
                              'nb_param': 1150852,
                              'Dataset':'AID',
                              'Train_size': 0.5,
                              'Remark': 'Kernel_size=(3,3)'}, ignore_index=True)
```

```python
df_result
```

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 22})
```

```python
df_result = df_result.sort_values('Train_size')
g_acc = df_result[df_result['Type']=='Graph']['Accuracy']
g_size = df_result[df_result['Type']=='Graph']['Train_size']
c_acc = df_result[df_result['Type']=='CNN']['Accuracy']
c_size = df_result[df_result['Type']=='CNN']['Train_size']

plt.figure(figsize=(20,10))
plt.plot(g_size, g_acc)
plt.plot(c_size, c_acc)
plt.legend(['Graph (987\'140 params)', 'CNN (1\'150\'852 params)'],
          loc='lower right')
plt.xlabel('Size of the training set')
plt.ylabel('Accuracy on test set')
plt.title('Accuracy vs dataset size')
plt.ylim(0, 100)
plt.savefig('Acc_vs_size.png')
plt.show()
```
